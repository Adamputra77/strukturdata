# Kode ini merupakan sebuah program sederhana untuk mengelola teks dengan fitur undo dan redo. Berikut adalah penjelasan singkat tentang bagaimana program tersebut bekerja



- Library yang Digunakan: Program menggunakan beberapa library standar C++ seperti iostream untuk input-output, stack untuk menyimpan teks yang dimasukkan, string untuk bekerja dengan string, dan limits untuk mengatur batas maksimum ukuran input.



- Makro CLEAR: Digunakan untuk membersihkan layar konsol, bergantung pada sistem operasi. Jika sistem operasi adalah Windows (_WIN32 atau _WIN64), maka CLEAR akan diatur sebagai "cls", jika tidak, maka diatur sebagai "clear".



- Fungsi KetikTeks: Menerima input teks dan menambahkannya ke dalam stack stackTeks. Saat teks baru dimasukkan, stack stackRedo dihapus, karena teks yang baru dimasukkan menghapus riwayat redo.

- Fungsi Undo dan Redo: Masing-masing digunakan untuk melakukan operasi undo dan redo. Untuk undo, teks terakhir dari stackTeks dipindahkan ke stackRedo, dan untuk redo, teks teratas dari stackRedo dipindahkan kembali ke stackTeks.

- Fungsi IsEmptyUndo dan IsEmptyRedo: Digunakan untuk memeriksa apakah stack stackTeks atau stackRedo kosong.



- Fungsi TampilkanMenu: Digunakan untuk menampilkan menu operasi yang tersedia kepada pengguna.



- Fungsi TungguDanLanjutkan: Digunakan untuk menunggu input dari pengguna sebelum melanjutkan ke langkah berikutnya. Hal ini membantu agar pengguna memiliki waktu untuk membaca output sebelum layar dikosongkan.



- Fungsi main: Mengatur alur utama program. Program akan terus menampilkan menu operasi, menerima input dari pengguna, dan mengeksekusi operasi yang dipilih hingga pengguna memilih untuk keluar (pilihan "7").



### Dengan demikian, program ini memberikan antarmuka sederhana untuk mengelola teks dengan fitur undo dan redo SEKIAN TERIMAKASIH
